```python
import json
```

```python
def process_samples():
    queries = []
    with open('/kaggle/input/bounding-boxes-ocr/input_train (1).jsonl', 'r') as json_file:
        for json_str in list(json_file):
            queries.append(json.loads(json_str))
    json_file.close()
    #print(queries[0])
    processed_queries = []
    for i in range(0, len(queries)):
        query = queries[i]
        new_query = '<image>'
        new_query = new_query + ' OCR: ' + query['OCR']
        new_query = new_query+' hero: '+str(query['hero'])
        new_query = new_query+' villain: '+str(query['villain'])
        new_query = new_query+' victim: '+str(query['victim'])
        new_query = new_query+' other: '+str(query['other'])
        new_query = new_query+ ' <|endofchunk|>'
        image = query['image']
        processed_queries.append((new_query, image))
    #print(processed_queries[0])
    return processed_queries
        
```

```python
import random
def create_examples(n, processed):
    #returns a chunk consisting of n examples
    images = [] #list of images 
    samples = '' #text to be inputted 
    rand_list = []
    for i in range(0, n):
        num = random.randint(0, len(processed)-1)
        samples = samples+' '+processed[num][0]
        images.append(processed[num][1])
    return images, samples 
    
```

```python
def create_queries():
    queries = []
    with open('/kaggle/input/bounding-boxes-ocr/input_val (1).jsonl', 'r') as json_file:
        for json_str in list(json_file):
            queries.append(json.loads(json_str))
    json_file.close()
    #print(queries[0])
    processed_queries = []
    for i in range(0, len(queries)):
        query = queries[i]
        new_query = '<image>'
        new_query = new_query + ' OCR: ' + query['OCR']
        new_query = new_query+' hero: '
        image = query['image']
        processed_queries.append((new_query, image))
    #print(processed_queries[0])
    return processed_queries
```

```python
processed_train = process_samples()
processed_queries = create_queries()
```

For now, we wont finetune the model and instead will use the pre-trained weights directly.

```python
from PIL import Image
import requests
import torch 
```

```python
def make_predictions(query, n, processed_train):
    images, examples = create_examples(n, processed_train)
    vision_x = []
    for i in range(0, n):
#         print(images[i])
        demo_image = Image.open(f"/kaggle/input/tututu/constrain22_dataset_covid19/images/{images[i]}")
        vision_x.append(image_processor(demo_image).unsqueeze(0))
    query_image = Image.open(f"/kaggle/input/tututu/constrain22_dataset_covid19/images/{query[1]}")
    vision_x.append(image_processor(query_image).unsqueeze(0))
    vision_x = torch.cat(vision_x, dim=0)
    vision_x = vision_x.unsqueeze(1).unsqueeze(0).to('cuda')
    tokenizer.padding_side = "left" # For generation padding tokens should be on the left
    text = examples+query[0]
    lang_x = tokenizer([text], return_tensors="pt", truncation=True, max_length = 2048,).to('cuda')
    generated_text = model.generate(vision_x=vision_x,lang_x=lang_x["input_ids"],attention_mask=lang_x["attention_mask"],max_new_tokens=20,num_beams=3,)
    return tokenizer.decode(generated_text[0])
    
```

```python
def convert_str_to_dict(string):
    open_para = []
    close_para = []
    for i in range(len(string)):
        if string[i]=='[':
            open_para.append(i)
        if string[i]==']':
            close_para.append(i+1)
    dict_hvv = {}
    keys = ['hero','villain','victim','other'] 
    for i in range(4) :
        dict_hvv[keys[i]] = set(eval(string[open_para[i]:close_para[i]+1]))
    return dict_hvv

convert_str_to_dict("hero: ['1', '2', '1'] villain ['2020'] victim: [] other: []")
```

```python
from tqdm import tqdm
predictions = []
for i in tqdm(range(0, len(processed_queries))): 
    query = processed_queries[i]
    pred = make_predictions(query, 4, processed_train)
    print(pred)
    #predictions.append(convert_str_to_dict(pred))
```
