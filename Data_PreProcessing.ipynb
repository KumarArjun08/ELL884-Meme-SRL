## Loading the YOLO Model

```python
from ultralytics import YOLO

# load a pretrained yolo model
# for image bounding boxes

model = YOLO("yolov8n.pt")  
```

```python
print(Image_list)
```

## Running the YOLO Model on our input images

```python
for img_path in Image_list:
    if img_path!='/kaggle/input/hero-villain-or-victim/test_data/test_images/{img_name}':
        
        # predict on an image
        results = model(img_path)
        sav_dir = '/kaggle/working/' + img_path[49:-4]

        # Save the bounded box images in the directory
        results[0].save_crop(save_dir = sav_dir, file_name='test')
```

## Loading the Image captioning pipeine

```python
from transformers import pipeline

pipe = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large", max_new_tokens = 30,device = 0)
```

## Running the pipeline on our all boxed images and concatenating to input


```python
from tqdm import tqdm
f = open('/kaggle/working/bounded_train_text.txt','a')

for dirname in tqdm(os.listdir('/kaggle/working/rain_images')):
    Full_Text = ''

    for dir_image in os.listdir('/kaggle/working/rain_images/'+dirname):
        for file in os.listdir('/kaggle/working/rain_images/'+dirname+'/'+dir_image):
    
            generated_text = pipe('/kaggle/working/rain_images/'+dirname+'/'+dir_image+'/'+file)[0]['generated_text']
            Full_Text = Full_Text + generated_text + '. '

    f.write(dirname+ ' : ' + Full_Text + '\n')
        
f.close()
```

## Outputing the transformed file

```python
f = open('/kaggle/input/bounded-test/Bounded_test.txt.txt','r')
data = f.readlines()
f.close()
```

```python
hashed_data = {}
for line in data:
    idx = line.index(':')-1
    hashed_data[line[:idx]] = line[idx+3:-2]
```

```python
import json
json_data_train = []
with open('/kaggle/input/hero-villain-or-victim/test_data/test_questions.jsonl', 'r') as json_file:
    for json_str in list(json_file):
        json_data_train.append(json.loads(json_str))
json_file.close()
```

```python
print(len(json_data_train))
```

```python
list_memes = []
for meme in json_data_train:
    text = meme['OCR']
    text_nobr = text.replace('\n',' ')
    image = meme['image'][:-4]
    dict_img_text = meme
    try:
        dict_img_text['OCR'] = text_nobr+hashed_data[image]
        list_memes.append(dict_img_text)
        print(dict_img_text)
    except KeyError:
        list_memes.append(dict_img_text)
print(len(list_memes))
        
with open("test.jsonl", 'a') as f:
    for item in list_memes:
        json.dump(item, f)
        f.write('\n')
```
